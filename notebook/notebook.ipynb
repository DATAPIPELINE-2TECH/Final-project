- We should now extract a table of states and regions in Brazil, a suggested solution was to extract it from the source table **brazil_covid19.csv**, we remark that each date describe the cases of 27 states which is the number of states in brazil

```python
brazil_regst = target[['state', 'region']].drop_duplicates()
```

### - Treatment

```python
print("W")
```

```python
info = data.merge(brazil_regst, how ='left', left_on='state', right_on='state')[target.columns[:5]]
```

```python
info.head(5)
```

```python
output = info.groupby(by=['date', 'region', 'state']).sum()
```

```python
output = output.sort_values(by=['date', 'region', 'state']).reset_index()#.head(5).to_csv('res.csv', index=False)
output
```

```python
output.head(100)
```

### - Output Data
